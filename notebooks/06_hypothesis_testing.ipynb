```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, norm
```

```python
#  In a toss of 4 coins what is the prob. of getting 2 heads P(X=2), and what is the prob.
# of getting 3 or less heads P(X<=3) hint = this is summing the P(X=0) + P(X=1) + P(X=2) + P(X=3)
# sample_space = [0,1,2,3,4]
n = 4 #4 cases
p = 0.5
results = list(range(n+1)) # Each independent case this = [0,1,2,3,4]
freqs = [binom.pmf(i,n,p) for i in results] # the arguments of binom.pmf (number of successes, events, prob)
plt.bar(results,freqs)
for i, val in enumerate(freqs): #Enumerate returns the index in the first temp variable, and the second temp variable is the value
    plt.text(i, val, round(val,2), ha="center") # the text arguments are the coordinates (x,y, and the value to print)
plt.grid(True)
plt.xlabel("number of heads")
plt.ylabel("probability")
plt.show()
```

```python
accum_freqs = [sum(freqs[:i+1]) for i in range(len(freqs))]
plt.bar(results, accum_freqs)
plt.grid(True)
plt.xlabel("number of heads")
plt.ylabel("probability")
plt.show()
```

```python
start = 0
end = 10
num = 100
numbers = [start+i*(end-start)/num for i in range(num)]
freqs = [norm.pdf(value,5,1) for value in numbers] # the arguments of norm.pdf (number of successes, medium, standard deviation)
plt.plot(numbers,freqs)
plt.grid(True)
plt.axvline(5, color = "red")
plt.axvline(4, color = "green")
plt.axvline(6, color = "green")
plt.axvline(3, color = "magenta")
plt.axvline(7, color = "magenta")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.show()
```

```python
accum_freqs = [norm.cdf(value,5,1) for value in numbers] # the arguments of norm.pdf (number of successes, medium, standard deviation)
plt.plot(numbers,accum_freqs)
plt.grid(True)
plt.xlabel("Values")
plt.ylabel("Accumulated Probability")
plt.show()
```

```python
# A machine produces cereal bags that weigh 3 kilos with a standard deviation of 0.5 kilos
# an operator gives us an observation saying that a machine isn't calibrated correctly
# a random sample [2.3, 2.9, 3.1, 2.5, 2.8, 3.0, 2.7] indicates that the machine definitely isn't producing 3 kilo bags
# is the operator right?
# Step 1 - Define the Hypothesis
# H0: mu = 3 kilos
# H1: mu != 3 kilos
# Level of Significance alpha = 0.05
# the given parameters are populations

sacks = np.array([2.3, 2.9, 3.1, 2.5, 2.8, 3.0, 2.7])
population_mean = 3
population_std = .5
sample_mean = sacks.mean()
sample_std = sacks.std()
print(f"mean = {sample_mean:.2f}")
print(f"standard deviation {sample_std:.2f}")
```

```python
z_score = (sample_mean-population_mean)/(population_std/np.sqrt(len(sacks))) 
print(f"z-score ={z_score:.2f}") # how much to the left in standard error
p_value = 2*(1-norm.cdf(abs(z_score)))
print(f"P-Value = {p_value:.2f}")
# if the p_value is greater than the level of significance we reject the alternative hypothesis
# if the p_value is lower than the level of significance then we don't have grounds to reject it
# then the observation from the operator is not valid
```
